* **İki ortaklama katmanı:** Görüntü boyutunu küçülterek modelin daha verimli çalışmasını sağlar.
* **Üç tam bağlantılı katman:** Görüntüdeki özelliklerin birbirleriyle olan ilişkilerini öğrenerek sınıflandırma yapar.
